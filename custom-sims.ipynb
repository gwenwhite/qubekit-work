```python
import planckton
from planckton.compounds import COMPOUND
from planckton.forcefields import FORCEFIELD
from planckton.init import Compound, Pack
from planckton.sim import Simulation
import unyt as u
import os
import foyer 
import parmed as pmd
import mbuild as mb
```

```python
y6 = "/Users/gwen/cmelab/repos/projects/qubekit-work/y6.mol2"
#bto = "/Users/gwen/cmelab/repos/projects/qubekit-work/bto.mol2"

ff_path = "/Users/gwen/cmelab/repos/projects/qubekit-work/gaff_custom.xml"
forcefield = foyer.Forcefield(forcefield_files=ff_path)
compound = Compound(y6)
packer = Pack(compound, n_compounds=50, density=0.3 * u.g / u.cm ** 3,
             ff=forcefield)
```

```python
system = packer.pack()
```

```python
#this is where you run your simulation
my_sim = Simulation(
    system,
    kT=[1.0],
    gsd_write=1e2,
    log_write=1e2,
    e_factor=1.0,
    n_steps=[1e4],
    mode="cpu",
    shrink_steps=1e5,
    tau = [1.0],
    target_length=packer.L,
    dt=0.0001
)
my_sim.run()
```
